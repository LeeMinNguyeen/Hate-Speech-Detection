```python
import json
from upstash_redis import Redis
```

```python
r = Redis(
  url=upstash_url,
  token=upstash_password
)
```

```python
keys = r.keys("*")  # Get all keys
print(keys)
```

```python
print(r.lrange(keys[0], 0, -1))
```

```python
survey_data = {key: r.lrange(key, 0, -1) for key in keys}
print(survey_data)
```

```python
from datetime import datetime

survey_json = {key: [] for key in keys}

for session, records in survey_data.items():
    for i, record in enumerate(records):
        record_dict = json.loads(record)
        if "timestamp" in record_dict:
            record_dict["timestamp"] = datetime.fromisoformat(record_dict["timestamp"]).isoformat()
        records[i] = json.dumps(record_dict)
    
    for record in records:
        survey_json[session].append(json.loads(record))

print(survey_json)
```

```python
cutoff_time = datetime.fromisoformat("2025-04-21T12:00:00+07:00")

filtered_survey_json = {}
for session, records in survey_json.items():
    include_session = True
    for record in records:
        if "timestamp" in record and datetime.fromisoformat(record["timestamp"]).astimezone(cutoff_time.tzinfo) < cutoff_time:
            include_session = False
            break
    if include_session:
        filtered_survey_json[session] = records

print(filtered_survey_json)
```

```python
key_count = len(filtered_survey_json)
print(f"Number of sessions: {key_count}")
```

```python
with open("all_sessions.json", "w", encoding="utf-8") as json_file:
    json.dump(filtered_survey_json, json_file, indent=4, ensure_ascii=False)
```

```python
finished_sessions = {}

for session, records in filtered_survey_json.items():
    has_start = any(isinstance(record, dict) and record.get("type") == "start_interview" for record in records)
    has_end = any(isinstance(record, dict) and record.get("type") == "end_interview" for record in records)
    if has_start and has_end:
        finished_sessions[session] = records

print(finished_sessions)
```

```python
key_count = len(finished_sessions)
print(f"Number of finished sessions: {key_count}")
```

```python
with open("finished_sessions.json", "w", encoding="utf-8") as json_file:
    json.dump(finished_sessions, json_file, indent=4, ensure_ascii=False)
```
